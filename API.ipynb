```python
# yelp has category of restaurants and they provide json file for that 
# I used a json file from the github post - https://gist.github.com/cristinecula/c7b1c6aa5ddb74789db9
with open('yelp_categories.json') as f:
    data = json.load(f)

restaurants = dict((k, v) for (k, v) in data.items() if v == 'restaurants')
restaurants
```

```python
# for key in restaurants.keys():
#     print(key)
```

```python
import time
yelp_restaurants = []
# Cycle through categories
for category in restaurants.keys():
    params['categories'] = category
    # Cycle through restaurants
    for offset_number in range(0,1000,50):
        params['offset'] = offset_number

        response = requests.get(url=URL, params=params, headers=headers)

        if not response.json().get('businesses', False):
            break

        yelp_restaurants.extend(response.json()['businesses'])

        print("{}: {}-{}".format(category, offset_number, offset_number+50))
        
        time.sleep(0.5) 
```

```python
Yelp_listing = [i for n, i in enumerate(yelp_restaurants) if i not in yelp_restaurants[n + 1:]]
```

```python
london_restaurants = sorted(Yelp_listing, key=lambda k: k['name'])
```

```python
#london_restaurants
```

## json to list 

```python
Business_ID = []
Bus_Name = []
Review_count = []
Rating =[]
Category_Alias = []
Category_Title = []
Lat = []
Long =[]
Price = []
Postcode = []


#for loop 
for r in london_restaurants:    
    try:
        Business_ID_1 = r['id']
    except:
        Business_ID_1 = None
    try:
        Bus_Name_1 = r['name']
    except:
        Bus_Name_1 = None
    try:
        Review_count_1 = r['review_count']
    except:
        Review_count_1 = None
    try:
        Rating_1 = r['rating']
    except:
        Rating_1 = None
    try:
        Category_Alias_1 = r['categories'][0]['alias']
    except:
        Category_Alias_1 = None
    try:
        Category_Title_1 = r['categories'][0]['title']
    except:
        Category_Title_1 = None
    try:
        Lat_1 = r['coordinates']['latitude']
    except:
        Lat_1 = None
    try:
        Long_1 = r['coordinates']['longitude']
    except:
        Long_1 = None
    try:
        Price_1 = r['price']#['price']
    except:
        Price_1 = None
    try:
        Postcode_1 = r['location']['zip_code']
    except:
        Postcode_1 = None
    #print(Category_Alias,Category_Title,Lat,Long,Postcode)
    
    Business_ID.append(Business_ID_1)
    Bus_Name.append(Bus_Name_1)
    Review_count.append(Review_count_1)
    Rating.append(Rating_1)
    Category_Alias.append(Category_Alias_1)
    Category_Title.append(Category_Title_1)
    Lat.append(Lat_1)
    Long.append(Long_1)
    Price.append(Price_1)
    Postcode.append(Postcode_1)
```

## List to DataFrame

```python
new_data = pd.DataFrame({ 'Business_ID':Business_ID ,
    'Bus_Name':Bus_Name ,
    'Review_count':Review_count ,
    'Rating':Rating ,
    'Category_Alias':Category_Alias ,
    'Category_Title':Category_Title ,
    'Lat':Lat ,
    'Long':Long ,
    'Price':Price ,
    'Postcode':Postcode 
})
```
