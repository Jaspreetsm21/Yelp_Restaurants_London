```python
data = pd.read_csv('all_yelp_listing.csv',index_col=0)
```

# missing values

```python
(data.isnull().sum().div(data.count())).mul(100)
```

There are couple of columns that have missing values. Lets check them one by one.

```python
#Category_Alias
data.columns
```

## Prices 

```python
data['Price '].value_counts()
```

```python
data['Price '] = data['Price '].fillna(0)
```

```python
data['Price '] = data['Price '].replace(to_replace=[0,'£','££','£££','££££'],value=[0,1,2,3,4])
```

```python
data['Price '].value_counts()
```

## Category 

```python
null_data = data[data.isnull().any(axis=1)]
null_data
```

```python
data['Category_Title '] = data['Category_Title '].fillna('Chinese')
data['Category_Alias '] = data['Category_Alias '].fillna('chinese')
```

```python
null_data = data[data.isnull().any(axis=1)]
null_data
```

## Lat and Long

There are 9 rows of lat/long missing data and I'm going to fillin the missing data with London's (51.5074° N, 0.1278° W)

```python
data['Lat '] = data['Lat '].fillna('51.5074')
data['Long '] = data['Long '].fillna('-0.1278')
```

## Postcode

```python
data['Postcode '] = data['Postcode '].fillna('W2 2SZ')
```
